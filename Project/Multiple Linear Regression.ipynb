```python
data['Extracurricular Activities']=data['Extracurricular Activities'].astype('category')
data.dtypes
```

```python
# Encoding
data['Extracurricular Activities']=data['Extracurricular Activities'].astype('category')
data['Extracurricular Activities']=data['Extracurricular Activities'].cat.codes
data
```

```python
data.info()
```

```python
# to find null value
data.isnull().sum()
```

```python
#feature
x = data.drop(columns='Performance Index')
x
```

```python
#Target
y = data['Performance Index']
y
```

```python
#splitting
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state = 0)
print(x_train.shape, x_test.shape,y_train.shape, y_test.shape)
```

```python
#modeling and validation
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
c = lr.intercept_
c
```

```python
m = lr.coef_
m
```

```python
#Evalution Metric for test data
from sklearn.metrics import r2_score
y_pred = lr.predict(x_test)
r2_score(y_test,y_pred)
```

```python
# realtionship between y_test and y_prediction
output = pd.DataFrame({'Actual_Performance Index' : y_test,'Predicted_Performance Index' :y_pred})
output.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred)
plt.xlabel('Actual_Performance Index')
plt.ylabel('Predicted_Performance Index')
plt.title('Testing Data ')
plt.show()
```

```python
#Evalution Metric for training data
from sklearn.metrics import r2_score
y_pred_train = lr.predict(x_train)
r2_score(y_train,y_pred_train)
```

```python
# realtionship between y_training and y_prediction
output = pd.DataFrame({'Actual_Performance Index' : y_train,'Predicted_Performance Index' :y_pred_train})
output.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_train,y_pred_train)
plt.xlabel('Actual_Performance Index')
plt.ylabel('Predicted_Performance Index')
plt.title('Traing Data')
plt.show()
```
